```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mesh_size = np.array([378 , 13934 , 50748 , 78344  , 102208 , 133964 , 163974 , 189226])
FEM_time  = np.array([0.012023 , 0.753185 , 3.58559 , 6.13537  , 8.76618 , 12.4143 , 16.7298 , 20.2124])
FEM_mem   = np.array([5.23651 , 293.555 , 1289.93 , 3443.85  , 4664.96 , 6342.31 , 8051.71, 9616.58])
julia_time= np.array([15.052167 , 746.073 , 3789.124 , 5530.508584  , 8168.542166 , 12206.497166 , 17576.8415 , 20042.462042])
julia_mem = np.array([17534.34765625 , 730279.08203125 , 3.30842173828125e6 , 5.42593022265625e6 , 7.34896924609375e6 , 1.026042169921875e7 , 1.283207200390625e7,1.538055710546875e7 ])
```

```python
julia_time = julia_time/1000
julia_mem = julia_mem /1024
```

```python
import matplotlib.pyplot as plt

# === Data arrays ===
# mesh_size, FEM_time, julia_time, FEM_mem, julia_mem
# should already be defined in your workspace

# === Create combined figure ===
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))  # 1 row, 2 columns

# --- Plot 1: Execution Time ---
ax1.plot(mesh_size, FEM_time, marker='o', label="FEM time")
ax1.plot(mesh_size, julia_time, marker='s', label="Julia time")
ax1.set_xlabel("Mesh size (elements)")
ax1.set_ylabel("Execution time (s)")
ax1.set_title("Solver Time vs. Mesh Size")
ax1.legend()
ax1.grid(True)

# --- Plot 2: Memory Usage ---
ax2.plot(mesh_size, FEM_mem, marker='o', linestyle='--', label="FEM memory")
ax2.plot(mesh_size, julia_mem, marker='s', linestyle='--', label="Julia memory")
ax2.set_xlabel("Mesh size (elements)")
ax2.set_ylabel("Peak memory (MiB)")
ax2.set_title("Solver Memory vs. Mesh Size")
ax2.legend()
ax2.grid(True)

# === Layout and save ===
fig.tight_layout()
fig.savefig("performance_comparison.png", dpi=300)
plt.show()
```
